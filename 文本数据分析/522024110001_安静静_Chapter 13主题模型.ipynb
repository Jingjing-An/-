```python
import pandas as pd

df=pd.read_excel('text_analysis_weibo_sample.xlsx')

print(df.head())
```

```python
import re
import jieba
from gensim import corpora,models
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
```

```python
def clean(text):
    # 该函数接受一个字符串参数 text
    nonums=re.sub('[^\u4e00-\u9fa5]','',text)
    
    # 使用 re.sub() 函数进行字符串替换操作。
    # '[^\u4e00-\u9fa5]' 是一个正则表达式，代表匹配所有非中文字符（中文字符的 Unicode 范围是 \u4e00 到 \u9fa5）。
    # 第二个参数 '' 表示将匹配到的非中文字符替换为空字符串。
    # text 是要进行替换操作的原始字符串。
    nonums=jieba.lcut(nonums)
    # 最终将替换后的字符串赋值给变量 nonums
    return nonums
    # 返回去除非中文字符后的字符串

test=df['标题/微博内容']
# 从数据框 df 中提取名为 '标题/微博内容' 的列，并将其赋值给变量 test
[clean(i) for i in test[:10]]
```

```python
df['微博内容分词']=df['标题/微博内容'].apply(clean)
df['微博内容分词']
```

```python
def clean_text(text):
    text = re.sub(r"[^\u4e00-\u9fa5]", "", text)
    words=jieba.lcut(text)
    stopwords = set(['的','了','这','为'])
    words = [word for word in words if word not in stopwords]
    text = ' '.join(words)
    return text

df['微博内容分词']=df['标题/微博内容'].apply(clean_text)
df['微博内容分词']=df['微博内容分词'].apply(lambda x: x.split())
```

```python
dictionary = corpora.Dictionary(df['微博内容分词'])     # 根据分词结果创建字典
corpus = [dictionary.doc2bow(text) for text in df['微博内容分词']]     # 根据分词结果创建语料库
lda_model = models.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)
```

```python
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```

```python
lda_vis = gensimvis.prepare(lda_model, corpus, dictionary)
```

```python
pyLDAvis.display(lda_vis)
```
